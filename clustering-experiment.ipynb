## Basic sanity checks

```python
import pprint
from IPython.display import Markdown, display

def printmd(string, color=None):
    colorstr = "<span style='color:{}'>{}</span>".format(color, string)
    display(Markdown(colorstr))
    
pp = pprint.PrettyPrinter(indent=4)

knobs = {}
data = {}

for r in results:
    knob_str = str(r[1])
    if not knob_str in knobs:        
        knobs[knob_str] = []
    if len(knobs[knob_str]) < 5000:
        knobs[knob_str].append(r[1])
    if not knob_str in data:        
        data[knob_str] = []
    if len(data[knob_str]) < 5000:
        data[knob_str].append(r[0])

printmd("**Found " + str(len(knobs)) + " different configurations in " + str(len(results)) + " data **", "green")

i = 1
for k in knobs:
    print "Configuration #" + str(i) + " (" + str(len(knobs[k])) + " data points):"
    i += 1
    pp.pprint(eval(k))
    print "\n"
```

## Basic plots

This is just for getting an idea of the distributions. 

```python
import matplotlib.pyplot as plt
import numpy
    
cars = 150.0

training_data = []
averages = []
percentiles_75 = []
percentiles_90 = []
st_deviations = []

for i in range(14):
    key = "{u'total_car_counter': " + str(cars) + "}"
    cars += 50

    dataset = [x["overhead"] for x in data[key]]
    
    plt.hist(dataset, bins=50, range=(1.0,4.0))  
    plt.show()
    
    avg = numpy.mean(dataset)
    med = numpy.median(dataset)
    per_75 = numpy.percentile(dataset, 75)
    per_90 = numpy.percentile(dataset, 90)
    std = numpy.std(dataset)
    var = numpy.var(dataset)
    
    averages.append([avg])
    percentiles_75.append([per_75])
    percentiles_90.append([per_90])
    st_deviations.append([std])

    training_data.append([avg, med, per_75, per_90, std, var])
    
    print key
    print "average: " + str(avg)
    print "median: " + str(med)
    print "75 percentile: " + str(per_75)
    print "90 percentile: " + str(per_90)
    print "standard deviation: " + str(std)
    print "variance: " + str(var)
    print "------------------------"
    
pp.pprint(training_data)
```

```python
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib as mpl
mpl.rc("savefig", dpi=150)

distortions = []
distortions2 = []
distortions3 = []

K = range(1,10)
for k in K:
    print "clusters: " + str(k)
    kmeans = KMeans(n_clusters=k, random_state=0).fit(training_data)
    print kmeans.labels_
    print kmeans.inertia_
    print kmeans.n_iter_
    print kmeans.cluster_centers_
    distortions.append(sum(np.min(cdist(training_data, kmeans.cluster_centers_, 'euclidean'), axis=1)) / len(training_data))
    distortions2.append(kmeans.inertia_)
    distortions3.append(kmeans.score(training_data))
    print "\n"

plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion as Euclidean distance')
plt.title('The Elbow Method showing the optimal k')
plt.show()

plt.plot(K, distortions2, 'bx-')
plt.ylabel('Distortion as Inertia')
plt.show()

plt.plot(K, distortions3, 'bx-')
plt.ylabel('Distortion as clustering score')
plt.show()
```
