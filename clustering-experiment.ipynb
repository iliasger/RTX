## Basic plots

This is just for getting an idea of the distributions. 

```python
import matplotlib.pyplot as plt
import numpy
    
cars = 150.0

training_data = []
averages = []
percentiles_75 = []
percentiles_90 = []
st_deviations = []

for i in range(14):
    key = "{u'total_car_counter': " + str(cars) + "}"
    cars += 50

    dataset = [x["overhead"] for x in data[key]]
    
    plt.hist(dataset, bins=50, range=(1.0,4.0))  
    plt.show()
    
    avg = numpy.mean(dataset)
    med = numpy.median(dataset)
    per_75 = numpy.percentile(dataset, 75)
    per_90 = numpy.percentile(dataset, 90)
    std = numpy.std(dataset)
    var = numpy.var(dataset)
    
    averages.append([avg])
    percentiles_75.append([per_75])
    percentiles_90.append([per_90])
    st_deviations.append([std])

    training_data.append([avg, med, per_75, per_90, std, var])
    
    print key
    print "average: " + str(avg)
    print "median: " + str(med)
    print "75 percentile: " + str(per_75)
    print "90 percentile: " + str(per_90)
    print "standard deviation: " + str(std)
    print "variance: " + str(var)
    print "------------------------"
    
pp.pprint(training_data)
```

```python
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib as mpl
mpl.rc("savefig", dpi=150)
from sklearn import metrics
from sklearn.metrics import pairwise_distances

distortions = []
distortions2 = []
distortions3 = []

K = range(1,10)
for k in K:
    print "clusters: " + str(k)
    kmeans = KMeans(n_clusters=k, random_state=0).fit(training_data)

    print kmeans.labels_
    print kmeans.inertia_
    print kmeans.n_iter_
    print kmeans.cluster_centers_
    distortion = sum(np.min(cdist(training_data, kmeans.cluster_centers_, 'euclidean'), axis=1)) / len(training_data)
    distortions.append(distortion)
    distortions2.append(kmeans.inertia_)
    distortions3.append(kmeans.score(training_data))
    print "\n"

plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion as Euclidean distance')
plt.title('The Elbow Method showing the optimal k')
plt.show()

plt.plot(K, distortions2, 'bx-')
plt.ylabel('Distortion as Inertia')
plt.show()

plt.plot(K, distortions3, 'bx-')
plt.ylabel('Distortion as clustering score')
plt.show()
```

## Sillouette analysis

```python
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib as mpl
mpl.rc("savefig", dpi=150)
from sklearn import metrics
from sklearn.metrics import pairwise_distances

silhouette_scores = []
cluster_numbers = range(2,10)
for k in cluster_numbers:
    print "clusters: " + str(k)
    
    kmeans = KMeans(n_clusters=k, random_state=0).fit(training_data)
    labels = kmeans.labels_
    print labels
    
    if k > 1:        
        silhouette_score = metrics.silhouette_score(training_data, labels, metric='euclidean')
        print silhouette_score
        silhouette_scores.append(silhouette_score)
    print "\n"
    
plt.plot(cluster_numbers, silhouette_scores, 'bo')
plt.xlabel('number of clusters')
plt.ylabel('Average Silhouette Score')
# plt.title('Title')
plt.show()
```

```python
from __future__ import print_function

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

for n_clusters in cluster_numbers:
    # Create a subplot with 1 row and 2 columns

    fig, ax1 = plt.subplots()
#     fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([0, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(training_data) + (n_clusters + 1) * 5])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(training_data)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(training_data, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(training_data, cluster_labels)

    y_lower = 5
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 5  # 10 for the 0 samples

    ax1.set_title("Silhouette plot for k = " + str(n_clusters))
    ax1.set_xlabel("Silhouette coefficient")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])

plt.show()
```
